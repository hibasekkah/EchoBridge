```python
import speech_recognition as sr
import pyttsx3
import customtkinter as ctk
import sqlite3
from tkinter import scrolledtext,PhotoImage
from PIL import Image, ImageTk



def record_audio():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Adjusting ambient noise... Please wait...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        print("You can speak now...")
        audio = recognizer.listen(source, timeout=20)
        print("Recording complete!")
    return audio

def recognize_text(audio):
    recognizer = sr.Recognizer()
    try:
        print("Recognizing text...")
        text = recognizer.recognize_sphinx(audio, language="en-US")
        print("You said: {}".format(text))
    except sr.UnknownValueError:
        print("Sorry, I could not understand the audio.")
    except sr.RequestError as e:
        print("Recognition error: {}".format(e))
    return text


def readtext(text):
    v = pyttsx3.init()
    voice = v.getProperty('voices')
    v.setProperty('voice', voice[1].id) 
    v.setProperty('rate', 150)  
    v.say(text)
    v.runAndWait()


    

ctk.set_appearance_mode("dark")

ctk.set_default_color_theme("blue")


app = ctk.CTk()
app.geometry("400x600")


conn = sqlite3.connect("Base_Donne.db")
cur = conn.cursor()
cur.execute("""
    CREATE TABLE IF NOT EXISTS USER (
        username TEXT PRIMARY KEY,
        password TEXT
    )
""")
conn.commit()
conn.close()
def main_window():
    app.title("SilentVoice")
    clear_window(app)
    label1 = ctk.CTkLabel(app, text="Register Here!", font=("Arial", 30))
    label1.grid(row=0, column=0, columnspan=2, pady=20, sticky="ew")

    button1 = ctk.CTkButton(app, text="Login", font=("Arial", 20), command=login)
    button1.grid(row=1, column=0, padx=20, pady=10, sticky="ew")

    button2 = ctk.CTkButton(app, text="Signup", font=("Arial", 20), command=signup)
    button2.grid(row=1, column=1, padx=20, pady=10, sticky="ew")

def clear_window(window):
    for widget in window.winfo_children():
        widget.destroy()
def login():
    def login_database():
        conn = sqlite3.connect("Base_Donne.db")
        cur = conn.cursor()
        username=username_entry.get().strip()
        password=password_entry.get().strip()
        if len(username)==0 or len(password)==0:
             result_label.configure(text_color="red", text="Please fill in the information carefully")
        else:
            cur.execute("SELECT * FROM USER WHERE username=? AND password=?", (username_entry.get(), password_entry.get()))
            row = cur.fetchone()
            if row:
                result_label.configure(text_color="white",text="Login successful!")
               
                app.after(1000, chat_window)
            else:
                result_label.configure(text_color="white",text="Login failed. \nUser not found or incorrect password.")
        conn.close()
    
    app.title("Login")
    clear_window(app)
    ctk.CTkButton(app, text="<", font=("Arial", 20,"bold"), command=main_window, width=30,corner_radius=0).grid(row=0, column=0,columnspan=1, padx=0, pady=0, sticky="wn")
    ctk.CTkLabel(app, text="Username:", font=("Arial", 20)).grid(row=1, column=0, padx=10, pady=10, sticky="ew")
    username_entry = ctk.CTkEntry(app,font=("Arial", 18))
    username_entry.grid(row=1, column=1, padx=10, pady=10, sticky="ew")

    ctk.CTkLabel(app, text="Password:", font=("Arial", 20)).grid(row=2, column=0, padx=10, pady=10, sticky="ew")
    password_entry = ctk.CTkEntry(app, show="*",font=("Arial", 20))
    password_entry.grid(row=2, column=1, padx=10, pady=10, sticky="ew")

    global result_label
    result_label = ctk.CTkLabel(app, text="Please fill in the information carefully", font=("Arial", 20))
    result_label.grid(row=3, column=0, columnspan=2, pady=10, sticky="ew")

    ctk.CTkButton(app, text="Login", font=("Arial", 20), command=login_database).grid(row=4, column=0, columnspan=2, padx=150, pady=10, sticky="ew")
    
    


def signup():
    def signup_database():
        conn = sqlite3.connect("Base_Donne.db")
        cur = conn.cursor()
        username=username_entry.get().strip()
        password=password_entry.get().strip()
        if len(username)==0 or len(password)==0:
             result_label.configure(text_color="red", text="Please fill in the information carefully")
        else:
            cur.execute("INSERT INTO USER (username, password) VALUES (?, ?)", (username_entry.get(), password_entry.get()))
            conn.commit()
            result_label.configure(text_color="white",text="Account created successfully")
        conn.close()

    app.title("Signup")
    clear_window(app)
    ctk.CTkButton(app, text="<", font=("Arial", 20,"bold"), command=main_window, width=30,corner_radius=0).grid(row=0, column=0,columnspan=1, padx=0, pady=0, sticky="wn")
    ctk.CTkLabel(app, text="Username:", font=("Arial", 20)).grid(row=1, column=0, padx=10, pady=10, sticky="ew")
    username_entry = ctk.CTkEntry(app,font=("Arial", 18))
    username_entry.grid(row=1, column=1, padx=10, pady=10, sticky="ew")

    ctk.CTkLabel(app, text="Password:", font=("Arial", 20)).grid(row=2, column=0, padx=10, pady=10, sticky="ew")
    password_entry = ctk.CTkEntry(app, show="*",font=("Arial", 20))
    password_entry.grid(row=2, column=1, padx=10, pady=10, sticky="ew")

    global result_label
    result_label = ctk.CTkLabel(app, text="Please fill in the information carefully", font=("Arial", 20))
    result_label.grid(row=3, column=0, columnspan=2, pady=10, sticky="ew")

    ctk.CTkButton(app, text="Sign Up", font=("Arial", 20), command=signup_database).grid(row=4, column=0, columnspan=2, padx=150, pady=10, sticky="ew")

def chat_window():
    app.title("Chat")
    clear_window(app)
    back_button = ctk.CTkButton(app, text="<", font=("Arial", 20,"bold"), command=login, width=30,corner_radius=0)
    back_button.pack(side='top', anchor='nw', padx=0, pady=0)
    
    chat_frame = ctk.CTkFrame(app)
    chat_frame.pack(expand=True, fill='both')
    chat_history = scrolledtext.ScrolledText(chat_frame, state='disabled', width=50, height=20, wrap='word')
    chat_history.pack(expand=True, fill='both', pady=10, padx=10)
    
    message_frame = ctk.CTkFrame(app)
    message_frame.pack(fill='x', pady=10, padx=10)
    
    message_entry = ctk.CTkEntry(message_frame)
    message_entry.pack(side='left', expand=True, fill='x', padx=5)

 
    send_icon_image = Image.open("send_icon.png")
    audio_icon_image = Image.open("audio_icon.png")
  
    send_icon = ctk.CTkImage(send_icon_image)
    audio_icon = ctk.CTkImage(audio_icon_image)
    
    def audio_text():
        audio=record_audio()
        if audio:
            message=recognize_text(audio)
            display_message("other", message, chat_history)
        
    def send_message():
        message = message_entry.get()
        if message.strip():
            message_entry.delete(0, 'end')
            display_message("me", message, chat_history)
    def display_message(sender, content, chat_history):
        chat_history.configure(state='normal')
        if sender=="me":
            audio_icon_black_image = Image.open("audio_icon_black.png")
            audio_icon_black = ctk.CTkImage(audio_icon_black_image)
            audio_button_black = ctk.CTkButton(chat_history, image=audio_icon_black, text="", command=lambda: readtext(content),hover_color="gray")
            audio_button_black.configure(fg_color="White")
    
            button_height = audio_button_black.winfo_height()
            audio_button_black.configure(width=button_height)
            chat_history.window_create('end', window=audio_button_black)
           
            label = ctk.CTkLabel(  chat_history,  text=content, text_color="white", fg_color="black", corner_radius=5, font=("Arial", 20), pady=10,padx=7, wraplength=300  )
            chat_history.insert('end', '\n')
            chat_history.window_create('end', window=label)
        elif sender=="other":
            label = ctk.CTkLabel(  chat_history,  text=content, text_color="white", fg_color="gray", corner_radius=5, font=("Arial", 20), pady=10,padx=7, wraplength=300  )
            chat_history.insert('end', '\n')
            chat_history.window_create('end', window=label)
    
        chat_history.insert('end', "\n")
        chat_history.insert('end', "\n")
        chat_history.configure(state='disabled')
        chat_history.see('end')
   
    audio_button = ctk.CTkButton(
        message_frame, 
        image=audio_icon, 
        command=audio_text, 
        text="",    
    )
    send_button = ctk.CTkButton(
        message_frame, 
        image=send_icon, 
        command=send_message, 
        text="", 
    )
    send_button.pack(side='right', padx=2)

    audio_button.pack(side='right')
    button_height = send_button.winfo_height()

    
    send_button.configure(width=button_height,corner_radius=5)
    audio_button.configure(width=button_height,corner_radius=5)
main_window()
app.grid_columnconfigure(0, weight=1)
app.grid_columnconfigure(1, weight=1)
app.grid_columnconfigure(2, weight=1)

app.mainloop()


```
